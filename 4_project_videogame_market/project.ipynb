```python
# Информация о пропусках
def show_missing_stats(tmp0):
    """
    Функция для отображения статистики пропущенных значений в DataFrame.
    """
    missing_stats = pd.DataFrame({
        'Кол-во пропусков': tmp0.isnull().sum(),
        'Доля пропусков': tmp0.isnull().mean()
    })
    missing_stats = missing_stats[missing_stats['Кол-во пропусков'] > 0]
    
    if missing_stats.empty:
        return "Пропусков в данных нет"
    
    # Форматируем при выводе через Styler
    return (missing_stats.style.format({'Доля пропусков': '{:.4f}'}).background_gradient(cmap='coolwarm'))
show_missing_stats(df_analysed)
```

```python
# Удаляем строки с пустыми значениями по полям 'year_of_release' и 'genre'
df_analysed = df_analysed.dropna(subset = ['year_of_release', 'genre'])
```

```python
# Заполним пропуски в данных по продажам с учетом платформы, жанра и региона (для каждого региона своя функция)
def fill_eu_sales(row):
    if pd.isna(row['eu_sales']):
        group = df_analysed[
            (df_analysed['platform'] == row['platform']) &
            (df_analysed['genre'] == row['genre'])
        ]
        return group['eu_sales'].mean()
    else:
        return row['eu_sales']
def fill_jp_sales(row):
    if pd.isna(row['jp_sales']):
        group = df_analysed[
            (df_analysed['platform'] == row['platform']) &
            (df_analysed['genre'] == row['genre'])
        ]
        return group['jp_sales'].mean()
    else:
        return row['jp_sales']
# Применим функции к датасету 
df_analysed['eu_sales'] = df_analysed.apply(fill_eu_sales, axis = 1)
df_analysed['jp_sales'] = df_analysed.apply(fill_jp_sales, axis = 1)
```

```python
# Заполним пропуски индикаторами (-1 или N/A)
df_analysed['critic_score'] = df_analysed['critic_score'].fillna(-1)
df_analysed['user_score'] = df_analysed['user_score'].fillna(-1)
df_analysed['rating'] = df_analysed['rating'].fillna('N/A')
```

## Категоризация данных
    
*Разделите все игры по оценкам пользователей и выделите такие категории: высокая оценка (от 8 до 10 включительно), средняя оценка (от 3 до 8, не включая правую границу интервала) и низкая оценка (от 0 до 3, не включая правую границу интервала).*

```python
# Создаем копию датасета
df_actual = df_actual.copy()
```

```python
# Создание функции для категоризации игр по оценкам пользователей
def category_user_score(user_score):
    if 8 <= user_score < 10:
        return 'Высокая оценка'
    elif 3 <= user_score < 8:
        return 'Средняя оценка'
    elif 0 <= user_score < 3:
        return 'Низкая оценка'
    else:
        return 'Оценка отсутствует'
# Применим созданную функцию к датасету
df_actual['user_score_category'] = df_actual['user_score'].apply(category_user_score)
```

```python
# Проверим создание новой колонки с категориями по оценкам пользователей
df_actual.head()
```

*Разделите все игры по оценкам критиков и выделите такие категории: высокая оценка (от 80 до 100 включительно), средняя оценка (от 30 до 80, не включая правую границу интервала) и низкая оценка (от 0 до 30, не включая правую границу интервала).*

```python
# Создание функции для категоризации игр по оценкам критиков
def category_critic_score(critic_score):
    if 80 <= critic_score < 100:
        return 'Высокая оценка'
    elif 30 <= critic_score < 80:
        return 'Средняя оценка'
    elif 0 <= critic_score < 30:
        return 'Низкая оценка'
    else:
        return 'Оценка отсутствует'
# Применим созданную функцию к датасету
df_actual['critic_score_category'] = df_actual['critic_score'].apply(category_critic_score)
```

```python
# Проверим создание новой колонки с категориями по оценкам критиков
df_actual.head()
```
